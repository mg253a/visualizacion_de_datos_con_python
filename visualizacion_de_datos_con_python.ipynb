# Visualización de datos con Python: Matplotlib, Seaborn y Plotly

# **Aula 1: Conociendo la biblioteca Matplolib**

```python
import pandas as pd

df = pd.read_csv('/content/inmigrantes_canada.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.set_index('Pais', inplace=True)
df
```

```python
anos = list(map(str, range(1980, 2014)))
anos
```

```python
colombia = df.loc['Colombia', anos]
colombia
```

```python
col_dict = {'Año': colombia.index.tolist(),
            'Inmigrantes': colombia.values.tolist()}
col_dict
```

```python
datos_col = pd.DataFrame(col_dict)
datos_col.tail()
```

Creando el primer gráfico con matplotlib

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(datos_col['Año'], datos_col['Inmigrantes'])
```

# **Aula 2: Creando figuras con Matplolib**

```python
plt.figure(figsize=(8,4))
plt.plot(datos_col['Año'], datos_col['Inmigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010','2014'])
plt.title('Inmigrantes colombianos hacia Canada')
plt.xlabel('Año')
plt.ylabel('No. de inmigrantes')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(datos_col['Año'], datos_col['Inmigrantes'])
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(datos_col['Año'], datos_col['Inmigrantes'])
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('Inmigrantes colombianos hacia Canada\n1980 - 2013')
ax.set_xlabel('Año')
ax.set_ylabel('No. de inmigrantes')
plt.show()
```

 Creando subplots en una dirección

```python
fig, axs = plt.subplots(1, 2, figsize=(10,4))

axs[0].plot(datos_col['Año'], datos_col['Inmigrantes'])
axs[0].set_title('Inmigrantes colombianos hacia Canada\n1980 - 2013')
axs[0].set_xlabel('Año')
axs[0].set_ylabel('No. de inmigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()

axs[1].boxplot(datos_col['Inmigrantes'])
axs[1].set_title('Boxplot de la inmigración de colombianos\n hacia Canadá 1980 - 2013')
axs[1].set_xlabel('Colombia')
axs[1].set_ylabel('No. de inmigrantes')
axs[1].grid()

plt.show()
```

```python
datos_col.describe()
```

 Creando subplots en dos direcciones #1

```python
fig, axs = plt.subplots(2,2, figsize=(10,6))
fig.subplots_adjust(hspace=0.5,wspace=0.3)

axs[0,0].plot(df.loc['Colombia', anos])
axs[0,0].set_title('Colombia')

axs[0,1].plot(df.loc['Brasil', anos])
axs[0,1].set_title('Brasil')

axs[1,0].plot(df.loc['Argentina', anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(df.loc['Perú', anos])
axs[1,1].set_title('Perú')

for ax in axs.flat:
  ax.set_xlabel('Año')
  ax.set_ylabel('No. de inmigrantes')
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.grid()

plt.show()
```

Creando subplots en dos direcciones #2

```python
fig, axs = plt.subplots(2,2, figsize=(10,6))
fig.subplots_adjust(hspace=0.5,wspace=0.3)
fig.suptitle('Inmigración de los 4 mayores países sudamericanos\n hacia Canada desde 1980 a 2013')

axs[0,0].plot(df.loc['Colombia', anos])
axs[0,0].set_title('Colombia')

axs[0,1].plot(df.loc['Brasil', anos])
axs[0,1].set_title('Brasil')

axs[1,0].plot(df.loc['Argentina', anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(df.loc['Perú', anos])
axs[1,1].set_title('Perú')

ymin = 0
ymax = 7000

for ax in axs.ravel():
  ax.set_ylim(ymin,ymax)

for ax in axs.flat:
  ax.set_xlabel('Año')
  ax.set_ylabel('No. de inmigrantes')
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.grid()

plt.show()
```

```python
colores = []
for pais in sudamerica_sorted.index:
  if pais == 'Colombia':
    colores.append('forestgreen')
  else:
    colores.append('gray')

fig, ax = plt.subplots(figsize=(10,3))

ax.barh(sudamerica_sorted.index, sudamerica_sorted['Total'], color=colores)
ax.set_title('Sudámerica: Colombia es el 2do país con más migrantes hacia\nCanadá durante el periodo 1980 - 2013', fontsize=16, loc='left')
ax.set_ylabel('')
ax.set_xlabel('No. de inmigrantes', fontsize=12)
ax.xaxis.set_tick_params(labelsize=10)
ax.yaxis.set_tick_params(labelsize=10)

plt.show()
```

```python
colores = []
for pais in sudamerica_sorted.index:
  if pais == 'Colombia':
    colores.append('forestgreen')
  else:
    colores.append('gray')

fig, ax = plt.subplots(figsize=(10,3))

ax.set_frame_on(False)
ax.xaxis.set_visible(False)
#ax.yaxis.set_visible(False)
ax.tick_params(axis='y', which='both', size=0)

ax.barh(sudamerica_sorted.index, sudamerica_sorted['Total'], color=colores)
ax.set_title('Sudámerica: Colombia es el 2do país con más migrantes hacia\nCanadá durante el periodo 1980 - 2013', fontsize=16, loc='left')
ax.set_ylabel('')
#ax.set_xlabel('No. de inmigrantes', fontsize=12)
#ax.xaxis.set_tick_params(labelsize=10)
ax.yaxis.set_tick_params(labelsize=10)

for i,j in enumerate(sudamerica_sorted['Total']):
  ax.text(j+20,i,str(j),color='black', fontsize=8.5, ha='left', va='center')

plt.show()
```

Personalizando las visualizaciones con Seaborn

```python
ax = sns.barplot(data=top_10, x='Total', y=top_10.index, orient='h')
ax.set(title='Top_10 países con mayor migración hacia Canadá\n1980 -2013',
       xlabel='No.de migrantes',
       ylabel='Año')

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4))

ax = sns.barplot(data=top_10, x='Total', y=top_10.index, orient='h')
ax.set_title('Top_10 países con mayor migración hacia Canadá\n1980 -2013', loc='left', fontsize=16)
ax.set_xlabel('No.de migrantes', fontsize=14)
ax.set_ylabel('')

plt.show()
```

Utilizando diversas palettes

```python
def generar_grafico(palette):

  fig, ax = plt.subplots(figsize=(8,4))
  ax.set_frame_on(False)
  ax.xaxis.set_visible(False)
  ax.tick_params(axis='y', which='both', size=0)
  ax = sns.barplot(data=top_10, x='Total',y=top_10.index, hue=top_10.index, orient='h',
                   palette=palette, legend=False)
  ax.set_title('Top_10 países con mayor migración hacia Canadá\n1980 -2013', loc='left', fontsize=16)
  ax.set_xlabel('No.de migrantes', fontsize=14)
  ax.set_ylabel('')
  for i,j in enumerate(top_10['Total']):
    ax.text(j+20,i,str(j),color='black', fontsize=8.5, ha='left', va='center')

plt.show()
```

Explorando otros temas

```python
def generar_grafico(palette):

  fig, ax = plt.subplots(figsize=(8,4))
  ax.set(xticklabels=[])
  ax = sns.barplot(data=top_10, x='Total',y=top_10.index, hue=top_10.index, orient='h',
                   palette=palette, legend=False)
  ax.set_title('Top_10 países con mayor migración hacia Canadá\n1980 -2013', loc='left', fontsize=16)
  ax.set_xlabel('No.de migrantes', fontsize=14)
  ax.set_ylabel('')
  sns.despine()
  for i,j in enumerate(top_10['Total']):
    ax.text(j+20,i,str(j),color='black', fontsize=8.5, ha='left', va='center')

plt.show()
```

```python
sns.set_theme()
generar_grafico('tab10')
```

```python
sns.set_theme(style='darkgrid')
generar_grafico('tab10')
```

```python
sns.set_theme(style='whitegrid')
generar_grafico('tab10')
```

```python
sns.set_theme(style='white')
generar_grafico('tab10')
```

# **Aula 5: Gráficos interactivos con Plotly**

 Creando nuestro primer gráfico interactivo

https://plotly.com/python/

```python
import plotly.express as px
```

```python
fig = px.line(datos_col, x='Año', y='Inmigrantes')
fig.update_layout(width=600, height=400, xaxis={'tickangle':-45})
fig.show()
```

Añadiendo rótulos y modificando fuentes

```python
fig = px.line(datos_col, x='Año', y='Inmigrantes',
              title='Inmigración de colombianos hacia Canadá entre 1980 y 2013')
fig.update_layout(width=600, height=400, xaxis={'tickangle':-45},
                  xaxis_title='Año', yaxis_title='No.de inmigrantes',
                  font_family='Arial', font_size=14, font_color='grey')
fig.show()
```

Alterando colores con Plotly

```python
fig = px.line(datos_col, x='Año', y='Inmigrantes',
              title='Inmigración de colombianos hacia Canadá entre 1980 y 2013')

fig.update_traces(line_color='green',line_width=3)
fig.update_layout(width=600, height=400, xaxis={'tickangle':-45},
                  xaxis_title='Año', yaxis_title='No.de inmigrantes',
                  font_family='Arial', font_size=14, font_color='grey')
fig.show()
```
