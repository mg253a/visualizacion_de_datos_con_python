# Visualización de datos con Python: Matplotlib, Seaborn y Plotly

# **Aula 1: Conociendo la biblioteca Matplolib**

```python
import pandas as pd

df = pd.read_csv('/content/inmigrantes_canada.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.set_index('Pais', inplace=True)
df
```

```python
anos = list(map(str, range(1980, 2014)))
anos
```

```python
colombia = df.loc['Colombia', anos]
colombia
```

```python
col_dict = {'Año': colombia.index.tolist(),
            'Inmigrantes': colombia.values.tolist()}
col_dict
```

```python
datos_col = pd.DataFrame(col_dict)
datos_col.tail()
```

Creando el primer gráfico con matplotlib

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(datos_col['Año'], datos_col['Inmigrantes'])
```

# **Aula 2: Creando figuras con Matplolib**

```python
plt.figure(figsize=(8,4))
plt.plot(datos_col['Año'], datos_col['Inmigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010','2014'])
plt.title('Inmigrantes colombianos hacia Canada')
plt.xlabel('Año')
plt.ylabel('No. de inmigrantes')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(datos_col['Año'], datos_col['Inmigrantes'])
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(datos_col['Año'], datos_col['Inmigrantes'])
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('Inmigrantes colombianos hacia Canada\n1980 - 2013')
ax.set_xlabel('Año')
ax.set_ylabel('No. de inmigrantes')
plt.show()
```

 Creando subplots en una dirección

```python
fig, axs = plt.subplots(1, 2, figsize=(10,4))

axs[0].plot(datos_col['Año'], datos_col['Inmigrantes'])
axs[0].set_title('Inmigrantes colombianos hacia Canada\n1980 - 2013')
axs[0].set_xlabel('Año')
axs[0].set_ylabel('No. de inmigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()

axs[1].boxplot(datos_col['Inmigrantes'])
axs[1].set_title('Boxplot de la inmigración de colombianos\n hacia Canadá 1980 - 2013')
axs[1].set_xlabel('Colombia')
axs[1].set_ylabel('No. de inmigrantes')
axs[1].grid()

plt.show()
```

 Creando subplots en dos direcciones #1

```python
fig, axs = plt.subplots(2,2, figsize=(10,6))
fig.subplots_adjust(hspace=0.5,wspace=0.3)

axs[0,0].plot(df.loc['Colombia', anos])
axs[0,0].set_title('Colombia')

axs[0,1].plot(df.loc['Brasil', anos])
axs[0,1].set_title('Brasil')

axs[1,0].plot(df.loc['Argentina', anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(df.loc['Perú', anos])
axs[1,1].set_title('Perú')

for ax in axs.flat:
  ax.set_xlabel('Año')
  ax.set_ylabel('No. de inmigrantes')
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.grid()

plt.show()
```

Creando subplots en dos direcciones #2

```python
fig, axs = plt.subplots(2,2, figsize=(10,6))
fig.subplots_adjust(hspace=0.5,wspace=0.3)
fig.suptitle('Inmigración de los 4 mayores países sudamericanos\n hacia Canada desde 1980 a 2013')

axs[0,0].plot(df.loc['Colombia', anos])
axs[0,0].set_title('Colombia')

axs[0,1].plot(df.loc['Brasil', anos])
axs[0,1].set_title('Brasil')

axs[1,0].plot(df.loc['Argentina', anos])
axs[1,0].set_title('Argentina')

axs[1,1].plot(df.loc['Perú', anos])
axs[1,1].set_title('Perú')

ymin = 0
ymax = 7000

for ax in axs.ravel():
  ax.set_ylim(ymin,ymax)

for ax in axs.flat:
  ax.set_xlabel('Año')
  ax.set_ylabel('No. de inmigrantes')
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.grid()

plt.show()
```

# **Aula 3: Personalizando gráficos con Matplotlib**

Personalizando visualizaciones con Matplotlib

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(datos_col['Año'], datos_col['Inmigrantes'], lw=4, marker='o')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('Inmigrantes colombianos hacia Canada\n1980 - 2013', fontsize=18, loc='left')
ax.set_xlabel('Año', fontsize=14)
ax.set_ylabel('No. de inmigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
ax.grid()
plt.show()
```

Alterando colores #1

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(datos_col['Año'], datos_col['Inmigrantes'], lw=2.5, color='g', marker='o')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title('Inmigrantes colombianos hacia Canada\n1980 - 2013', fontsize=18, loc='left')
ax.set_xlabel('Año', fontsize=14)
ax.set_ylabel('No. de inmigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
ax.grid()
plt.show()
```

```python
sudamerica = df.query('Region == "América del Sur"')
sudamerica.head(10)
```

```python
colores = ['royalblue','orange','forestgreen','orchid',
           'purple','brown','slateblue','gray','olive',
           'navy','teal','tomato']

fig, ax = plt.subplots(figsize=(10,3))

ax.bar(sudamerica.index, sudamerica['Total'], color=colores)
ax.set_title('Inmigración de sudamericanos hacia Canadá\ndurante el periodo 1980 - 2013', fontsize=18, loc='left')
ax.set_xlabel('')
ax.set_ylabel('No. de inmigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

plt.show()

```

 Alterando colores #2

```python
colores = ['royalblue','orange','forestgreen','orchid',
           'purple','brown','slateblue','gray','olive',
           'navy','teal','tomato']

fig, ax = plt.subplots(figsize=(10,3))

ax.barh(sudamerica.index, sudamerica['Total'], color=colores)
ax.set_title('Inmigración de sudamericanos hacia Canadá\ndurante el periodo 1980 - 2013', fontsize=18, loc='left')
ax.set_ylabel('')
ax.set_xlabel('No. de inmigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

plt.show()
```

```python
sudamerica_sorted = sudamerica.sort_values('Total', ascending=True)

colores = ['royalblue','orange','forestgreen','orchid',
           'purple','brown','slateblue','gray','olive',
           'navy','teal','tomato']

fig, ax = plt.subplots(figsize=(10,3))

ax.barh(sudamerica_sorted.index, sudamerica_sorted['Total'], color=colores)
ax.set_title('Inmigración de sudamericanos hacia Canadá\ndurante el periodo 1980 - 2013', fontsize=18, loc='left')
ax.set_ylabel('')
ax.set_xlabel('No. de inmigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

plt.show()
```

```python
colores = []
for pais in sudamerica_sorted.index:
  if pais == 'Colombia':
    colores.append('forestgreen')
  else:
    colores.append('gray')

fig, ax = plt.subplots(figsize=(10,3))

ax.barh(sudamerica_sorted.index, sudamerica_sorted['Total'], color=colores)
ax.set_title('Sudámerica: Colombia es el 2do país con más migrantes hacia\nCanadá durante el periodo 1980 - 2013', fontsize=16, loc='left')
ax.set_ylabel('')
ax.set_xlabel('No. de inmigrantes', fontsize=12)
ax.xaxis.set_tick_params(labelsize=10)
ax.yaxis.set_tick_params(labelsize=10)

plt.show()
```
